# A Python Integer Is More Than Just an Integer

El INTEGER (entero) de Python es un puntero a una posición en la memoria que contiene toda la información del objeto, incluidos los bytes que contienen el valor entero. Esta información adicional en la estructura INTEGER de Python es lo que permite que Python se programa de forma libre y dinámica. Toda esta información adicional en los tipos Python tiene un "costo" que se hace especialmente evidente en estructuras que combinan muchos de estos objetos.

# A Python List Is More Than Just a List

Consideremos ahora qué sucede cuando usamos una estructura de datos de Python que contiene muchos objetos de Python. El contenedor multielemento mutable estándar en Python es la LISTA.

```python
L = list(range(10)) #Creo una lista de enteros
```

```python
L
```

```python
type(L[0]) #Verifico que el tipo de dato de la lista sean enteros
```

```python
L2 = [str(c) for c in L] #Creo un array de stings tomando como referencia la lista de enteros L
```

```python
L2
```

```python
type (L2[0]) #Verifico que el tipo de dato de la lista sean strings
```

```python
L3 = [True,"2",3.0,4] #Debido a su tipado dinamico (no necesito declarar el tipo de dato) es que puedo crear listas heterogeneas
```

```python
L3
```

```python
[type(item)for item in L3] #Armo una lista que contenga los tipos de datos de cada uno de los componentes de la lista L3
```

Como se observa del caso anterior las listas de Python son muy flexibles ya que permiten contener elementos de diferentes tipos de datos

# Fixed-Type Arrays in Python

Python ofrece diferentes opciones para almacenar datos en forma eficiente. Esto son "almacenadores" de datos fijos

```python
#Array de Python (disponible desde la version de Python 3.3)

import array #Importo el modulo "array"
L = list(range(10)) #Creo una lista de 10 elementos 
A = array.array("i",L) #Utilizo la lista para crear un array
A
```

En el caso anterior la creacion del array se realizo con la sentencia array.array(typecode[, initializer])

typecode: A string with all available type codes. En este ejemplo typecode="i"



Mucho mas eficiente es el objeto ndarray del paquete NumPy.

    * El objeto "array" provee una forma eficiente de almacenamiento de datos
    * El objeto "ndarray" de Numpy tiene una forma eficiente de operaciones con los datos, por lo tanto para estas operaciones se prefiere por sobre el array

```python
import numpy as np #Convencion estandar para importar NumPy utilizada por la comunidad
```

Creacion de Arrays from Python Lists

```python
np.array([1,4,2,5,3]) #Creo un array en forma manual
```

```python
#Los arrays de NumPy estan RESTRINGIDOS a que todos los elementos de su contenido SEAN DEL MISMO TIPO DE OBJETO
#En caso que los tipos de objetos NO coincidan NumPy tratara de uniformizar dentro de la estructura jerarquica de los objetos

np.array([3.14,4,2,3]) #En este caso todos los elementos se convierten a float
```

```python
#Si queremos definir en forma EXPLICITA el tipo de dato que debe tener el array se puede definir con dtype

np.array([1,2,3,4],dtype="float32") #Defino explicitamente que los elementos del array sean "float32"
```

```python
#Los arrays de NumPy pueden ser explicitamente multidimensionales}
#Una forma de inicializar un array multidimensional utilizando una lista es:

np.array([range(i,i+3)for i in[2,4,6]])
```

Creacion de arrays "from Scratch"

```python
#Especialmente para arrays de gran dimension es mas eficiente crear arrays "from Scratch" utilizando rutinas construidas dentro de NumPy
```

```python
#Creo un array 1D de zeros con tipo de dato int

np.zeros(10,dtype=int)
```

```python
#Creo un array 2D de 3x5 con tipo de dato float

np.ones((3,5),dtype=float)
```

```python
#Creo un array 2D de 3x5 y le asigno el valor de 3.14 a todos los elementos

np.full((3,5),3.14)
```

```python
#Creo un array 1D equiespaciado de numeros enteros comenzando en 0, terminando en 20 (sin incluir) y step=2

np.arange(0,20,2)
```

```python
#Creo un array 1D con valores entre 0 y 1, con 5 valores equiespaciados entre ellos

np.linspace(0,1,5)
```

```python
#Creo un array 2D de 3x3 de numeros aleatorios distribucion uniforme
#Valores entre 0 y 1

np.random.random((3,3))
```

```python
#Creo un array 2D de 3x3 de numeros aleatorios con distribucion normal
#Valores con media(mean)=0 y desviacion estandar(standard deviation)=1

np.random.normal(0,1,(3,3))
```

```python
#Creo un array 2D de 3x3 de numeros aleatorios enteros
#Valores entre 0 y 10

np.random.randint(0,10,(3,3))
```

```python
#Creo una matriz identidad 2D de 3x3

np.eye(3)
```

```python
#Creo un array "vacia" la cual contendra los valores sin inicializar.
#Los valores que se muentran seran aquellos que existan en ese lugar de memoria en ese momento
#Es importante destacar que NO SIEMPRE los valores seran ceros

np.empty(3)
```
